```python
# Import libraries
%matplotlib inline 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cleaned up Dataset location
fileName = "https://library.startlearninglabs.uw.edu/DATASCI410/Datasets/EnergyEfficiencyData.csv"
df = pd.read_csv(fileName)
```

### Preliminary Findings (i.e. first glance)

#### Heating load is strongly and positively correlated with Cooling load.

```python
sns.set_style("whitegrid") # setting seaborn style
sns.lmplot(x='Heating Load', y='Cooling Load', data=df) # seaborn lmplot of Heating Load v Cooling Load
plt.title('Heating Load v. Cooling Load') # title
plt.show() # display without object information
'Correlation: {}'.format(df['Heating Load'].corr(df['Cooling Load'])) # printing pearson correlation coefficient between Loads.
```

While not entirely identical (cooling load tends to be slightly higher in comparison to heating load), heating load and cooling load largely behave similarly given other house features. 

#### Surface Area and Relative Compactness have a strong negative correlation

```python
sns.lmplot(x='Surface Area', y='Relative Compactness', data=df)
plt.title('Surface Area v. Relative Compactness')
plt.show()
'Correlation: {}'.format(df['Surface Area'].corr(df['Relative Compactness']))
```

While not a perfectly linear relationship, Surface Area and Relative Compactness are likely derived from the same house dimensions. 

## Main Findings

### 1. Overall Height has a strong positive correlation with Heating and Cooling Load

```python
fig = plt.figure(figsize=(6,6)) # creating figure
ax = fig.gca() # obtaining axes
sns.violinplot(x = 'Overall Height', y = 'Heating Load', data=df, ax = ax) # creating seaborn violin plot 
ax.set_title('Violin Plot of Overall Height and Heating Load\nof 768 Homes') # title
ax.set_xlabel('Overall Height') #x label
ax.set_ylabel('Heating Load') #y lable
plt.show() # displaying plot without additional object info
'Correlation: {}'.format(df['Overall Height'].corr(df['Heating Load'])) # displaying Pearson Correlation Coefficient
```

The violin plot above shows the greatly varying heating load distributions for houses with a height of 3.5m compared with those with a height of 7m. This and the high positive correlation between Overall Height and Heating Load shows that height is a key factor for predicting a house's heating load.

#### A. Relative Compactness has a moderate negative correlation with both Heating and Cooling load when Overall Height is taken into account

### 3. Window Area, a created feature, has a very strong correlation with Heating and Cooling Loads when Relative Compactness is taken into account.

```python
df['Window Area'] = df['Glazing Area'] * df['Wall Area'] # Window Area is the product Glazing Area and Wall Area
sns.lmplot(x='Window Area', y='Heating Load', data=df, hue=df.columns[0]) # Relative Compactness as hue 
plt.title('Window Area v. Heating Load\nSplit by Relative Compactness') # title
plt.show() # displaying plot without object information

for i in df['Relative Compactness'].unique(): # A for loop printing correlation coefficients for individual lines. 
    corr = df.loc[df.loc[:, 'Relative Compactness'] == i, 'Window Area'].corr( 
        df.loc[df.loc[:, 'Relative Compactness'] == i, 'Heating Load']) # obtaining pearson correlation coefficient
    print('Correlation between Window Area and Heating Load for Homes with {} Relative Compactness : {}'.format(i, corr)) # print
```

Window Area, the product of Glazing Area (a ratio) and Wall Area (m^2), factored by Relative Compactness, generally has a high positive correlation with Heating Load. Pearon coefficients for relative compactness groups range from 0.888 (0.71 Relative Compactness) to 0.956 (0.98 Relative Compactness). It would appear that Window Area and Relative Compactness are highly accurate predictors of Heating Load when both are taken into account.

```python
sns.lmplot(x='Window Area', y='Cooling Load', data=df, hue=df.columns[0])
plt.title('Window Area v. Cooling Load\nSplit by Relative Compactness')
plt.show()

for i in df['Relative Compactness'].unique():
    corr = df.loc[df.loc[:, 'Relative Compactness'] == i, 'Window Area'].corr(
        df.loc[df.loc[:, 'Relative Compactness'] == i, 'Cooling Load'])
    print('Correlation between Window Area and Cooling Load for Homes with {} Relative Compactness : {}'.format(i, corr))
```

Window Area, the product of Glazing Area (a ratio) and Wall Area (m^2), factored by Relative Compactness, generally has a high positive correlation with Cooling Load. Pearon coefficients for relative compactness groups range from 0.463 (0.79 Relative Compactness) to 0.986 (0.98 Relative Compactness). It would appear that Window Area and Relative Compactness are highly accurate predictors of Cooling Load when both are taken into account, though they are generally more accurate at predicting Heating Load.

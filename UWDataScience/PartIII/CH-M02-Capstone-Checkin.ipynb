## Project Goal
This project looked at four Energy Star benchmarking datasets: 3 Seattle-based datasets (2015, 2016, and 2017) and one Chicago-based (2014-2016) dataset. Created by EPA's ENERGY STAR, the Energy Star score is a rating (1-100) that compares the energy comsumption of similar buildings across the U.S. Along with Energy Star scores, the datasets are comprised of various building attributes concerning energy and property use.

There are five goals for this project, the first three of which will be largely achieved with this Milestone:
<ol>
    <li>Determine which features are most important for achieving a high Energy Star score.</li>
    <li>Determine whether the year recordings were taken (i.e. time) changes which features are salient in the best model by comparing Seattle 2015, 2016, and 2017 datasets.</li>
    <li>Determine whether the city where recordings were taken (i.e. location) changes which features are salient in the best model by comparing Seattle 15-17 and Chicago datasets.</li>
    <li>Develop a regression model that accurately predicts Energy Star scores.</li>
    <li>Develop a classification model that accurately predicts an Energy Star score's quintile.</li>
</ol>

## Data Sources

The Seattle dataset is the aggregate of three separate datasets, distinguised by the year recordings were taken (2015, 2016, and 2017). The 2015 and 2016 Seattle datasets were obtained from Kaggle ([Source](https://www.kaggle.com/city-of-seattle/sea-building-energy-benchmarking)) and were provided as part of a mandate the Seattle City Council passed in 2016, making the data for all buildings 20,000 square feet and larger available annually. Reportedly, both datasets represent nearly a third of Seattle's annual emissions. The 2017 dataset, released this November, was obtained directly from the Seattle Open Data Portal ([Source](https://data.seattle.gov/dataset/2017-Building-Energy-Benchmarking/qxjw-iwsh/)).


The Chicago dataset was obtained from the US goverment's open data website ([Source](https://catalog.data.gov/dataset/chicago-energy-benchmarking)) and was similarly provided through a law passed by Chicago city council, wherein buildings greater than 50,000 square feet are to track and annually report whole-building energy use. Reportedly, this dataset represents a fifth of Chicago's annual emissions.

## Methodology
As our goals consist of creating both a classification model that accurately predicts an Energy Star score's quintile as well as a regression model that accurately predicts Energy Star scores, we will be looking at both multi-classifcation and regression models. 

For multi-classification, either a decision tree (DT), k-neighbors (KNN), naive-bayes (GNB), or support-vectors classifier (SVC) could be the most performant. 

For regression, a linear model (e.g. ridge, lasso, elastic-net) would likely be most performant, though k-neighbors (KNN), decision tree (DTR), and support vector regressors (SVR) should also be tested.

### Pruning features based on correlations

```python
# Plotting df_seattle correlations as a heatmap
plt.figure(figsize=(48,42))
sns.heatmap(df_seattle.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap of Seattle Dataset Correlations', fontdict={'fontsize': 30})
plt.show()
```

#### Figure 1. Heatmap of Seattle Dataset Correlations
Before engineering and selecting features, a visual scan for highly correlative features was conducted. SiteEUI(kBtu/sf), SourceEUI(kBtu/sf), and SiteEnergyUse(kBtu) were all removed due to their representation within SiteEUIWN(kBtu/sf), SourceEUIWN(kBtu/sf), SiteEnergyUseWN(kBtu) respectively. SteamUse(kBtu) was also removed as it was shown to be highly correlative with DataYear2017 in comparison to 2015 or 2016; this could suggest an error or a change in recording practices between 2015-2016 and 2017. Finally, both Electricity(kWh) and NaturalGas(therms) were both represented as Electricity(kBtu) and NaturalGas(kBtu), and therefore were removed as well. 

As the BuildingType features are all parent-classes for PrimaryPropertyType features, they were removed due to the risk for colinearity, especially in the case of Low-Rise, Mid-Rise, and High-Rise property categories. For the rest of the one-hot encoded categorical features, one feature was removed per feature group: PrimaryPropertyType_Worship Facility for PrimaryPropertyType and CouncilDistrictCode_1 for CouncilDistrictCode.

```python
df_seattle.drop(['SiteEUI(kBtu/sf)', 'SourceEUI(kBtu/sf)', 'SiteEnergyUse(kBtu)',
                 'SteamUse(kBtu)', 'Electricity(kWh)', 'NaturalGas(therms)',
                 'BuildingType_Campus', 'BuildingType_Multifamily HR (10+)',
                 'BuildingType_Multifamily LR (1-4)', 'BuildingType_Multifamily MR (5-9)',
                 'BuildingType_NonResidential', 'BuildingType_Nonresidential COS',
                 'BuildingType_SPS-District K-12', 'PrimaryPropertyType_Worship Facility',
                 'CouncilDistrictCode_1'], axis=1, inplace=True)
```

#### Figure 2. Heatmap of Chicago Dataset Correlations
As with the Seattle dataset, several features were removed from the Chicago dataset prior to any feature engineering and selection. SiteEUI(kBtu/sf) and SourceEUI(kBtu/sf) were both removed due to their representation within SiteEUIWN(kBtu/sf) and SourceEUIWN(kBtu/sf) respectively. Likwise, all BuildingType features were removed due to high colinearity with PrimaryPropertyType features. For other one-hot encoded categorical features, PrimaryPropertyType_Worship Facility was removed from PrimaryPropertyType.

```python
# Feature Engineering
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Creating BuildingAvgGFA: average total square feet per building for a given property
df_sea_pre.loc[df_sea_pre['NumberofBuildings'] == 0, 'NumberofBuildings'] = 1
df_sea_pre['BuildingAvgGFA'] = df_sea_pre['PropertyGFABuilding(s)'] \
                             / df_sea_pre['NumberofBuildings']
df_chi_pre['BuildingAvgGFA'] = df_chi_pre['PropertyGFATotal'] \
                             / df_chi_pre['NumberofBuildings']
df_seattle['BuildingAvgGFA'] = \
    scaler.fit_transform(df_sea_pre['BuildingAvgGFA'].values.reshape(-1, 1))
df_chicago['BuildingAvgGFA'] = \
    scaler.fit_transform(df_chi_pre['BuildingAvgGFA'].values.reshape(-1, 1))

# Creating BuildingAvgElectric: average annual electricity usage (kBtu) per building
df_sea_pre['BuildingAvgElectric'] = df_sea_pre['Electricity(kBtu)'] \
                                  / df_sea_pre['NumberofBuildings']
df_chi_pre['BuildingAvgElectric'] = df_chi_pre['Electricity(kBtu)'] \
                                  / df_chi_pre['NumberofBuildings']
df_seattle['BuildingAvgElectric'] = \
    scaler.fit_transform(df_sea_pre['BuildingAvgElectric'].values.reshape(-1, 1))
df_chicago['BuildingAvgElectric'] = \
    scaler.fit_transform(df_chi_pre['BuildingAvgElectric'].values.reshape(-1, 1))

# Creating ParkingBuildingGFARatio: ratio of parking space v. building floor area
df_sea_pre['ParkingBuildingGFARatio'] = df_sea_pre['PropertyGFAParking'] \
                                      / df_sea_pre['PropertyGFABuilding(s)']
df_seattle['ParkingBuildingGFARatio'] = \
    scaler.fit_transform(df_sea_pre['ParkingBuildingGFARatio'].values.reshape(-1, 1))

# Creating SiteSourceEUIWNRatio: ratio of site v. source energy use (weather normalized)
df_sea_pre['SiteSourceEUIWNRatio'] = df_sea_pre['SiteEUIWN(kBtu/sf)'] \
                                   / df_sea_pre['SourceEUIWN(kBtu/sf)']
df_chi_pre['SiteSourceEUIWNRatio'] = df_chi_pre['SiteEUIWN(kBtu/sf)'] \
                                   / df_chi_pre['SourceEUIWN(kBtu/sf)']
df_seattle['SiteSourceEUIWNRatio'] = \
    scaler.fit_transform(df_sea_pre['SiteSourceEUIWNRatio'].values.reshape(-1, 1))
df_chicago['SiteSourceEUIWNRatio'] = \
    scaler.fit_transform(df_chi_pre['SiteSourceEUIWNRatio'].values.reshape(-1, 1))
```

```python
# Splitting datasets into training and testing sets, stratified by ENERGYSTARQuintile
from sklearn.model_selection import train_test_split

def TrainTestPrint(features, target, prefix, suffix):
    '''Splits arrays into training and testing; 
    returns list of arrays and shape of each array
    '''
    x_train, x_test,\
    y_train, y_test = train_test_split(features, target,
                                       test_size=0.2, random_state=42,
                                       stratify=target)
    print(f'x_{prefix}_train_{suffix}: {x_train.shape} |'
          f' x_{prefix}_test_{suffix}: {x_test.shape}\n'
          f'y_{prefix}_train_{suffix}: {y_train.shape} |'
          f' y_{prefix}_test_{suffix}: {y_test.shape}\n')
    return [x_train, x_test, y_train, y_test]

cls_sea = TrainTestPrint(x_sea, y_cls_sea, 'cls', 'sea')
cls_sea15 = TrainTestPrint(x_sea15, y_cls_sea15, 'cls', 'sea15')
cls_sea16 = TrainTestPrint(x_sea16, y_cls_sea16, 'cls', 'sea16')
cls_sea17 = TrainTestPrint(x_sea17, y_cls_sea17, 'cls', 'sea17')
cls_chi = TrainTestPrint(x_chi, y_cls_chi, 'cls', 'chi')
```

```python
# Classification: Selecting features when target is ENERGYSTARQuintile using SelectKBest
from sklearn.feature_selection import SelectKBest, f_classif

x_train_cls = [cls_sea[0], cls_sea15[0], cls_sea16[0], cls_sea17[0], cls_chi[0]]
y_train_cls = [cls_sea[2], cls_sea15[2], cls_sea16[2], cls_sea17[2], cls_chi[2]]
names = ['Seattle 2015-2017', 'Seattle 2015', 'Seattle 2016', 'Seattle 2017', 'Chicago 2014-2016']

dfcls_feature_select = pd.DataFrame()
for i in range(len(x_train_cls)):
    dataset = x_train_cls[i]
    # Dropping columns with zero variance
    for column in dataset.columns:
        if len(dataset[column].value_counts()) == 1:
            dataset = dataset.drop(column, axis=1)
    # Iterating over SelectKBest for top 30 ordered list of features indices
    col_nums = []
    for k in range(1, 31): 
        test = SelectKBest(score_func=f_classif, k=k)
        fit = test.fit(dataset, y_train_cls[i])
        new_col_nums = test.get_support(indices=True)
        new_col_num = list(set(new_col_nums) - set(col_nums))[0]
        col_nums.append(new_col_num)
    # Obtaining list of feature names
    features = []
    for col_num in col_nums:
        features.append(dataset.columns[col_num])
    dfcls_feature_select[names[i]] = features

print('Classification: Top 30 features for Seattle and Chicago Datasets')
dfcls_feature_select
```

```python
# Regression: Selecting features when target is ENERGYSTARScore using SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs

x_train_reg = [reg_sea[0], reg_sea15[0], reg_sea16[0], reg_sea17[0], reg_chi[0]]
y_train_reg = [reg_sea[2], reg_sea15[2], reg_sea16[2], reg_sea17[2], reg_chi[2]]
names = ['Seattle 2015-2017', 'Seattle 2015', 'Seattle 2016', 'Seattle 2017', 'Chicago 2014-2016']

sfs_models = []
for i in range(len(x_train_reg)):
    dataset = x_train_reg[i]
    target = y_train_reg[i]
    # Dropping columns with zero variance
    for column in dataset.columns:
        if len(dataset[column].value_counts()) == 1:
            dataset = dataset.drop(column, axis=1)
    # Instantiating SequentialFeatureSelector
    lr = LinearRegression()
    sfs = SFS(lr, k_features=len(dataset.columns)-2, #Dropping final 2 features for better visual range 
              forward=True, floating=False,
              scoring='neg_mean_squared_error', cv=3)
    sfs = sfs.fit(dataset, target)
    # Creating SFS plots
    fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')
    fig.set_size_inches(12,5)
    plt.title(f'{names[i]} dataset: Sequential Forward Selection (w. StdErr)')
    plt.show()
    # Obtaining data for all models
    models = sfs.get_metric_dict()
    sfs_models.append(models)
```

```python
# Testing CART, KNN, GNB, and SVC classifiers on 5 datasets with 6 combinations of features
from sklearn.model_selection import KFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

x_test_cls = [cls_sea[1], cls_sea15[1], cls_sea16[1], cls_sea17[1], cls_chi[1]]
y_train_cls = [cls_sea[2], cls_sea15[2], cls_sea16[2], cls_sea17[2], cls_chi[2]]
y_test_cls = [cls_sea[3], cls_sea15[3], cls_sea16[3], cls_sea17[3], cls_chi[3]]

cls_feature_sets = [x_train_cls, cls30, cls20, cls10, clsnum, clscat]

classifiers = [DecisionTreeClassifier(), KNeighborsClassifier(),
               GaussianNB(), SVC(gamma='auto')]

kfold = KFold(n_splits=5, random_state=42)

# Obtaining 5-fold CV training set score and test set score for classification results
rows = []
for i in range(len(datasets)):
    for j in range(len(cls_feature_sets)):
        row = []
        for k in range(len(classifiers)):
            # Obtaining score of 5-fold CV of training set
            results = cross_val_score(classifiers[k], cls_feature_sets[j][i],
                                      y_train_cls[i], scoring='accuracy', cv=kfold)
            # Obtaining score of test set after fitting to training set
            model = classifiers[k].fit(cls_feature_sets[j][i], y_train_cls[i])
            y_predict = model.predict(x_test_cls[i][cls_feature_sets[j][i].columns])
            acc_score = accuracy_score(y_test_cls[i], y_predict) * 100
            # Appending both results to row
            row.append(f'{acc_score:0.2f}%')
            row.append(f'{results.mean()*100:0.2f}% (+/-{results.std()*100:0.2f}%)')
        rows.append(row)
classification_results = pd.DataFrame(rows, index=multiindexrow, columns=multiindexcol)
classification_results
```

```python
# Testing Linear, Ridge, Lasso, Elastic, KNR, DTR, SVR regressors
# on 5 datasets with 5 combinations of features
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error 

x_test_reg = [reg_sea[1], reg_sea15[1], reg_sea16[1], reg_sea17[1], reg_chi[1]]
y_train_reg = [reg_sea[2], reg_sea15[2], reg_sea16[2], reg_sea17[2], reg_chi[2]]
y_test_reg = [reg_sea[3], reg_sea15[3], reg_sea16[3], reg_sea17[3], reg_chi[3]]

reg_feature_sets = [x_train_reg, regtop, regnum, regcat, reg30c] 

regressors = [LinearRegression(), Ridge(), Lasso(), ElasticNet(),
              KNeighborsRegressor(), DecisionTreeRegressor(), SVR(gamma='auto')]

kfold = KFold(n_splits=5, random_state=42)

# Obtaining 5-fold CV training set score and test set score for regression results
rows = []
for i in range(len(x_train_reg)):
    for j in range(len(reg_feature_sets)):
        row = []
        for k in range(len(regressors)):
            # Obtaining neg. MSE score of 5-fold CV of training set
            results = cross_val_score(regressors[k], reg_feature_sets[j][i],
                                      y_train_reg[i], scoring='neg_mean_squared_error', cv=kfold)
            # Obtaining neg. MSE score of test set after fitting to training set
            model = regressors[k].fit(reg_feature_sets[j][i], y_train_reg[i])
            y_predict = model.predict(x_test_reg[i][reg_feature_sets[j][i].columns])
            n_mse = mean_squared_error(y_test_reg[i], y_predict) * -1.0
            # Appending both results to row
            row.append(f'{n_mse:0.2f}')
            if results.mean() < - 9999:
                row.append('X')
            else:
                row.append(f'{results.mean():0.2f} (+/-{results.std():0.2f})')
        rows.append(row)
regression_results = pd.DataFrame(rows, index=multiindexrow, columns=multiindexcol)
print('')
regression_results
```

### Model Performance Evaluation

```python
def BestModelsPlotter(results):
    ''' 
    Obtains best scoring model per classifier/regressor per dataset 
    from the given feature sets with results dataframe. Returns barplots and df.
    '''
    # Obtaining ordered lists of models, datasets, and feature_sets from df
    models_unordered = list(results.columns.levels[0])
    models_order = [x for i, x in enumerate(results.columns.labels[0])
                    if x not in results.columns.labels[0][0:i]]
    models = [models_unordered[i] for i in models_order]
    datasets_unordered = list(results.index.levels[0])
    datasets_order = [x for i, x in enumerate(results.index.labels[0])
                      if x not in results.index.labels[0][0:i]]
    datasets = [datasets_unordered[i] for i in datasets_order]
    feature_sets_unordered = list(results.index.levels[1])
    feature_sets_order = [x for i, x in enumerate(results.index.labels[1])
                          if x not in results.index.labels[1][0:i]]
    feature_sets = [feature_sets_unordered[i] for i in feature_sets_order]
    scoring = results.columns.names[1]
    
    # Obtaining best scoring model out of the feature sets per model per dataset
    best_dataset_models = []
    for dataset in datasets:
        for model in models:
            tests = []
            dataset_model = results.loc[dataset, model].values
            for row in dataset_model:
                if row[0][-1] == '%': # Classification
                    tests.append(float(row[0][:-1]))
                else: # Regression
                    tests.append(float(row[0]))
            best_score = max(tests)
            best_features = feature_sets[tests.index(best_score)]
            best_dataset_model = (best_score, dataset, model, best_features)
            best_dataset_models.append(best_dataset_model)
    df_models = pd.DataFrame(best_dataset_models,
                             columns=[scoring, 'Dataset', 'Model', 'Feature Set'])
    # Plotting scores grouped by Model (ax1) and Feature Set (ax2)
    plt.figure(figsize=(20,8))
    ax1 = plt.subplot(121)
    ax1 = sns.barplot(x=scoring, y='Dataset', hue='Model',
                      data=df_models, hue_order=models,
                      orient='h', palette="gist_earth")
    ax2 = plt.subplot(122)
    ax2 = sns.barplot(x=scoring, y='Dataset', hue='Feature Set',
                      data=df_models, hue_order=feature_sets,
                      orient='h', palette="gist_earth")
    plt.suptitle('Best Scores per Model per Dataset, grouped by Model (left) and Feature Set (right)',
                 fontsize=16, y=0.94)
    plt.show()
    # Returning best models as a dataframe, to accompany barplots
    results = df_models.T
    results.columns = ['Set ' + str(i) for i in range(len(df_models.T.columns))]
    return results
```
